```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Visualization mile per gallon 


### Load the data


```python
sns.set()
```

```python
mpg = sns.load_dataset('mpg')
```

```python
mpg.head()
```

### Analyze dataset

```python
sns.lmplot(x='displacement', y='horsepower', data=mpg.head(100))
```

```python
sns.barplot(x='displacement', y='weight', data=mpg.head(100))
```

### This visuals shows that as much horsepower increase the displacement of pistons will increase with it.
