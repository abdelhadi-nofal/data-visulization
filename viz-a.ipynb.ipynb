```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Visualization For Tips

### Load the data

```python
sns.set()
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

### Analyze dataset

```python
sns.lmplot(x='total_bill', y='tip', data=tips.head(100))
```

```python
sns.barplot(x='total_bill', y='tip', data=tips.head(50))
```

### This visuals shows that as much total bill increase the tips increase with it.
